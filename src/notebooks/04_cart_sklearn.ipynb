## Classification Tree from SKLearn

```python
from sklearn.datasets import load_iris
from sklearn import tree
import sys
sys.path.append("..")
import data_processing.train_test_split as split
iris = load_iris()
x,y = iris.data, iris.target
cart = tree.DecisionTreeClassifier()
cart.fit(x,y)
```

```python
tree.plot_tree(cart)
```

```python
x_train, x_test, y_train, y_test = split.train_test_split(x,y, seed=2108)
cart.fit(x_train, y_train)
tree.plot_tree(cart)
```

```python
cart.predict(x_test)
```

```python
import numpy as np
print(np.mean(cart.predict(x_test) == y_test))
print(cart.predict(x_test) == y_test)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
tree.plot_tree(cart, filled=True, feature_names=iris.feature_names, class_names=list(iris.target_names))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
conf_mat = confusion_matrix(y_test, cart.predict(x_test))

labels = iris.target_names
disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat, display_labels=labels)
disp.plot(cmap=plt.cm.Oranges)  
plt.show()
```
